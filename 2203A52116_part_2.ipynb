```python
import pandas as pd
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
healthcare_data = {
    'Person': people,
    'Age': age,
    'Height': height,
    'Weight': weight
}
healthcare_df = pd.DataFrame(healthcare_data)
print(healthcare_df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(healthcare_df['Age'], healthcare_df['Height'])
plt.title('Scatter Plot: Age vs. Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

```python
plt.bar(healthcare_df['Person'], healthcare_df['Weight'])
plt.title('Bar Chart: Person vs. Weight')
plt.xlabel('Person')
plt.ylabel('Weight')
plt.show()
```

```python
plt.hist(healthcare_df['Age'], bins=10, color='blue', edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(healthcare_df['Age'], healthcare_df['Height'])
plt.title('Scatter Plot: Age vs. Height')
plt.xlabel('Age')
plt.ylabel('Height (in cm)')
plt.grid(True)
plt.show()
```

```python
plt.bar(healthcare_df['Person'], healthcare_df['Weight'])
plt.title('Bar Chart: Person vs. Weight')
plt.xlabel('Person')
plt.ylabel('Weight (in kg)')
plt.grid(axis='y')
plt.show()
```
